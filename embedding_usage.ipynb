```python
table_definitions = {
    "purchase_orders": """
        Table purchase_orders:
        - id:  primary key
        - supplier_id:  foreign key to suppliers
        - order_date: Date of purchase order
        - status: Status of order 'Ordered', 'ASN', 'Received'
    """,
    "purchase_order_lines": """
        Table purchase_order_lines:
        - id:  primary key
        - purchase_order_id: foreign key to purchase_orders
        - product_id: foreign key to products
        - quantity: Quantity of product
        - unit_price: Item price for each unit
    """,
    "sales_orders": """
        Table sales_orders:
        - id: primary key
        - customer_id: foreign key to customers
        - order_date: Date of sales order
        - status: Status of order 'Ordered', 'Shipped', 'Delivered'
        - shipping_address: Address of shipping
        - billing_address: Address of billing
    """,
    "sales_order_lines": """
        Table sales_order_lines:
        - id: primary key
        - sales_order_id: foreign key to sales_orders
        - product_id: foreign key to products
        - quantity: Quantity of product
        - unit_price: Unit price for each unit
    """
}

table_names = list(table_definitions.keys())
texts = list(table_definitions.values())

print(table_names)
print(texts)
```

```python
from sentence_transformers import SentenceTransformer
from transformers import AutoTokenizer, AutoModel
import torch
import numpy as np

embedding_models = {
    "MiniLM": SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2"),
    "E5": SentenceTransformer("intfloat/e5-base-v2"),
    "BGE": SentenceTransformer("BAAI/bge-base-en-v1.5"),
    "GTE": SentenceTransformer("thenlper/gte-base")
}
```

```python
import pandas as pd

simple_text = "This is a sample text to show how each ebmedding is generating the vector"
print("Embedding dimensionality for each model:\n")

for model_name, model in embedding_models.items():
    if "E5" in model_name:
        text_input = "query:" + simple_text
    else:
        text_input = simple_text

    embedding = model.encode([text_input], convert_to_numpy=True)
    print(f"{model_name}: {embedding.shape[1]} dimensions")
```

```python
import faiss

index_store = {}

for model_name, model in embedding_models.items():
    print(f"Encoding with {model_name}...")
    embeddings = model.encode(texts, convert_to_numpy=True, normalize_embeddings=True)

    index = faiss.IndexFlatIP(embeddings.shape[1])  # using cosine similarity
    index.add(embeddings)

    index_store[model_name] = {
        "index": index,
        "model": model
    }
```

```python
query = "get sales order ?"

print(f"\nQuery: {query}\n")

for model_name, data in index_store.items():
    model = data["model"]
    index = data["index"]

    if "E5" in model_name:
        query_input = "query: " + query
    else:
        query_input = query

    query_vector = model.encode([query_input], convert_to_numpy=True, normalize_embeddings=True)
    D, I = index.search(query_vector, k=1)

    match_name = table_names[I[0][0]]
    match_score = D[0][0]

    print(f"{model_name}: Top match = {match_name} (Score: {match_score:.4f})")
```
